### The percentage of Null Values in Each column

```python
def drop_columns(data, percent):
    percentage = data.isna().sum() / data.shape[0] * 100
    nullPerc = pd.DataFrame(percentage.sort_values(ascending=False)).reset_index()
    nullPerc.columns = ['Cols', 'Percentage']
    colmns = nullPerc[nullPerc['Percentage'] >= percent]
    dropped = colmns['Cols'].values
    print('Dropped Columns:', dropped)
    data.drop(columns=dropped,axis = 1, inplace=True)
    return None

    
```

### Cheking the columns with more than 50% null values

```python
above50 = drop_columns(data, 50)
print(above50)
```

### Drop these columns as we can't impute the null values

```python
above20 = drop_columns(data, 20)
print(above20)
```

### Fill latitude and longitude from city and vice versa

```python
data = data.sort_index()
```

```python
from geopy.geocoders import Nominatim
inst = Nominatim(user_agent ='mostafa') 
```

```python
data['city'].value_counts().sort_values(ascending = False)
```

```python

for index , row in data.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        if row['city'] == 'Unknown' or pd.isna(row['city']):
            try: 
                response = inst.reverse(f"{row['latitude']},{row['longitude']}" , language='en')
                #print(response.raw['address'])
                city = response.raw['address']
                if 'city' in city.keys(): 
                    data.at[index, 'city'] = city['city']
                    print(city['city'])
                elif 'town' in city.keys():
                    data.at[index, 'city'] = city['town']
                    print(city['town'])
                elif 'village' in city.keys():
                    data.at[index, 'city'] = city['village']
                    print(city['village'])
                elif 'county' in city.keys():
                    data.at[index, 'city'] = city['county']
                    print(city['county'])
                elif 'state' in city.keys():
                    data.at[index, 'city'] = city['state']
                    print(city['state'])
                else:
                    data.at[index, 'city'] = city['country']
                    print(city['country'])
            except:
                continue
        else:    
            continue
    else:
        continue

```
